```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0 , 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, label='Sine Wave', color='blue')
plt.plot(x, y2, label='coSine Wave', color='red')
plt.title('Line Plot - Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
categories = ['A', 'B', 'C']
values = [10, 15, 7]
plt.bar(categories, values, color='pink')
plt.title('barplot')
plt.xlabel('category')
plt.ylabel('values')
plt.show()
```

```python
import seaborn as sns
import numpy as np
data = np.random.normal(size=1000)
sns.histplot(data, bins=50, kde=True, color='red')
plt.title('histogram example')
plt.show()
```

```python
data = np.random.normal(size=100)
plt.boxplot(data)
plt.title( 'Box Plot of data')
plt.ylabel('data')
plt.show()
```

```python
import pandas as pd
data = pd.DataFrame({ 
    'Age': [25, 45, 34, 42, 25, 36, 52, 39, 40, 30],
    'Salary': [40000, 80000, 60000, 85000, 53000, 62000, 70000, 383989, 339893, 39839]
})
sns.scatterplot(x= data['Age'], y=data['Salary'], color='purple')
plt.title('Age vs Salary')
plt.show()
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)  # (rows, columns, position) 
sns.scatterplot(x=data['Age'], y=data['Salary'], color='purple')
plt.title('Age vs Salary (Scatter Plot)')
plt.xlabel('Age')
plt.ylabel('Salary')

# Histogram (2nd subplot) 
plt.subplot(1, 3, 2)  # (rows, columns, position) 
sns.histplot(data['Age'], bins=5, kde=True, color='green') 
plt.title('Histogram of Age') 
plt.xlabel('Age') 
plt.ylabel('Frequency') 

# Box Plot (3rd subplot) 
plt.subplot(1, 3, 3)  # (rows, columns, position) 
sns.boxplot(data['Salary'], color='orange') 
plt.title('Box Plot of Salary') 
plt.xlabel('Salary') 
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd  
iris = sns.load_dataset('iris')
print(iris.head())
```

```python
plt.figure(figsize=(8,6)) 
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', 
hue='species', style='species')
plt.title('Sepal Length vs Petal Length') 
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
# Histogram of sepal length 
plt.figure(figsize=(8,6 )) 
sns.histplot(iris[ 'sepal_length' ], bins=20, kde=True)
plt.title('Distribution of Sepal Length') 
plt.xlabel( 'Sepal Length (cm)') 
plt.ylabel
plt.show()
```

```python
# Pair plot of the Iris dataset 
sns.pairplot(iris, hue= 'species')
plt.show()
```
